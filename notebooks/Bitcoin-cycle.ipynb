```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create single plot
fig, ax1 = plt.subplots(figsize=(15, 8))

# Plot price trends by halving cycle
for cycle in sorted(df['Halving'].unique()):
    cycle_data = df[df['Halving'] == cycle]
    ax1.plot(cycle_data['Timestamp'], cycle_data['Close'], 
             label=f'Halving {cycle}', alpha=0.8)

ax1.set_title('Bitcoin Price During Different Halving Cycles')
ax1.set_xlabel('Date')
ax1.set_ylabel('Price (USD)')
ax1.legend()
# ax1.set_yscale('log')  # Using log scale since price changes are exponential
ax1.grid(True)

plt.tight_layout()
plt.show()

# Alternative visualization: Normalized price changes
plt.figure(figsize=(15, 8))

for cycle in sorted(df['Halving'].unique()):
    cycle_data = df[df['Halving'] == cycle]
    # Normalize prices to start at 100 for each cycle
    normalized_prices = cycle_data['Close'] / cycle_data['Close'].iloc[0] * 100
    # Create relative timeindex (days from start of cycle)
    days_from_start = (cycle_data['Timestamp'] - cycle_data['Timestamp'].iloc[0]).dt.total_seconds() / (24*60*60)
    
    plt.plot(days_from_start, normalized_prices, label=f'Halving {cycle}', alpha=0.8)

plt.title('Normalized Bitcoin Price Changes by Halving Cycle')
plt.xlabel('Days from Start of Cycle')
plt.ylabel('Price (Normalized to 100)')
plt.legend()
plt.yscale('log')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Define start dates for each halving cycle
halving_start_dates = {
    1: pd.to_datetime('2012-11-28 15:24:37'),
    2: pd.to_datetime('2016-07-09 16:46:13'),
    3: pd.to_datetime('2020-05-11 19:23:43'),
    4: pd.to_datetime('2024-04-20 12:09:27')  # Example date for the fourth cycle
}

# Calculate the 200th and 400th day for each cycle
halving_date_ranges = {}
for cycle, start_date in halving_start_dates.items():
    day_200 = start_date + pd.Timedelta(days=200)
    day_400 = start_date + pd.Timedelta(days=400)
    halving_date_ranges[cycle] = (day_200, day_400)

# Print the first and last date of the date range for each halving cycle
for cycle, (day_200, day_400) in halving_date_ranges.items():
    print(f"Cycle {cycle}:")
    print(f"  Start of 200-day range: {day_200.strftime('%Y-%m-%d')}")
    print(f"  End of 400-day range: {day_400.strftime('%Y-%m-%d')}\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame with the 'Timestamp' and 'Halving' columns

# Define target dates
negative_events = [
    pd.to_datetime('2013-12-05'),  # China ban
    pd.to_datetime('2013-10-02'),  # Silk Road shutdown
    pd.to_datetime('2013-06-16'),  # Mt. Gox halted USD withdrawals
    pd.to_datetime('2013-08-05'),  # Mt. Gox announced losses
    pd.to_datetime('2017-08-01'),  # bitcoin hard fork
    pd.to_datetime('2017-03-10'),  # SEC rejects the winklevoss bitcoin etf
    pd.to_datetime('2017-05-12'),  # wannacry ransomwhere attack
    pd.to_datetime('2021-05-12'),  # Elon Musk u-turn on bitcoin
    pd.to_datetime('2021-05-19'),  # Market crash?
]

positive_events = [
    pd.to_datetime('2013-12-11'),  # Microsoft accepts bitcoin
    pd.to_datetime('2013-10-29'),   # First bitcoin ATM installed
    pd.to_datetime('2013-07-03'),   # EOS and Tezos launches
    pd.to_datetime('2017-07-21'),   #BIP-91 acitivated
    pd.to_datetime('2021-02-08'),  # Tesla buys bitcoin
    pd.to_datetime('2021-04-14'),  # Coinbase IPO
    pd.to_datetime('2021-06-09'),  # el salvador adopts bitcoin
]

plt.figure(figsize=(15, 8))

for cycle in sorted(df['Halving'].unique()):
    cycle_data = df[df['Halving'] == cycle]
    days_from_start = (cycle_data['Timestamp'] - cycle_data['Timestamp'].iloc[0]).dt.total_seconds() / (24*60*60)
    
    # Filter for days between 200 and 400
    mask = (days_from_start >= 200) & (days_from_start <= 400)
    filtered_days = days_from_start[mask]
    
    normalized_prices = cycle_data['Close'] / cycle_data['Close'].iloc[0] * 100
    filtered_prices = normalized_prices[mask]
    
    plt.plot(filtered_days, filtered_prices, label=f'Halving {cycle}', alpha=1.0)
        
    # Plot markers for negative events (red)
    for target_date in negative_events:
        event_mask = (cycle_data['Timestamp'].dt.date == target_date.date()) & mask
        if any(event_mask):
            event_day = filtered_days[event_mask].iloc[0]
            event_price = filtered_prices[event_mask].iloc[0]
            plt.plot(event_day, event_price, 'ro', markersize=5)  # Red dot
        
    # Plot markers for positive events (green)
    for target_date in positive_events:
        event_mask = (cycle_data['Timestamp'].dt.date == target_date.date()) & mask
        if any(event_mask):
            event_day = filtered_days[event_mask].iloc[0]
            event_price = filtered_prices[event_mask].iloc[0]
            plt.plot(event_day, event_price, 'go', markersize=5)  # Green dot

plt.title('Normalized Bitcoin Price Changes by Halving Cycle (Days 200-400)')
plt.xlabel('Days from Start of Cycle')
plt.ylabel('Price (Normalized to 100)')
plt.legend()
plt.yscale('log')
plt.grid(True)
plt.show()
```
