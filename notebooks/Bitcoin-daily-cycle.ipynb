```python
# aggregating by days

import pandas as pd

# Assuming df is your original DataFrame with a 'Timestamp' column in datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='s')  # Ensure Timestamp is in datetime format

# Set the 'Timestamp' as the index
df.set_index('Timestamp', inplace=True)

# Resample the data to daily frequency and aggregate
daily_df = df.resample('D').agg({
    'Open': 'first',   # First value of the day
    'High': 'max',     # Maximum value of the day
    'Low': 'min',      # Minimum value of the day
    'Close': 'last',   # Last value of the day
    'Volume': 'sum',    # Sum of the volume for the day
    'Halving': 'max'
})

# Reset the index if you want 'Timestamp' as a column again
daily_df.reset_index(inplace=True)

# Display the first few rows of the new daily DataFrame
print(daily_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create single plot
fig, ax1 = plt.subplots(figsize=(15, 8))

# Plot price trends by halving cycle
for cycle in sorted(daily_df['Halving'].unique()):
    cycle_data = daily_df[daily_df['Halving'] == cycle]
    ax1.plot(cycle_data['Timestamp'], cycle_data['Close'], 
             label=f'Halving {cycle}', alpha=0.8)

ax1.set_title('Bitcoin Price During Different Halving Cycles')
ax1.set_xlabel('Date')
ax1.set_ylabel('Price (USD)')
ax1.legend()
# ax1.set_yscale('log')  # Using log scale since price changes are exponential
ax1.grid(True)

plt.tight_layout()
plt.show()

# Alternative visualization: Normalized price changes
plt.figure(figsize=(15, 8))

for cycle in sorted(daily_df['Halving'].unique()):
    cycle_data = daily_df[daily_df['Halving'] == cycle]
    # Normalize prices to start at 100 for each cycle
    normalized_prices = cycle_data['Close'] / cycle_data['Close'].iloc[0] * 100
    # Create relative timeindex (days from start of cycle)
    days_from_start = (cycle_data['Timestamp'] - cycle_data['Timestamp'].iloc[0]).dt.total_seconds() / (24*60*60)
    
    plt.plot(days_from_start, normalized_prices, label=f'Halving {cycle}', alpha=0.8)

plt.title('Normalized Bitcoin Price Changes by Halving Cycle')
plt.xlabel('Days from Start of Cycle')
plt.ylabel('Price (Normalized to 100)')
plt.legend()
plt.yscale('log')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# halving dates
halving_dates = [
    pd.to_datetime('2012-11-28 15:24:37'),
    pd.to_datetime('2016-07-09 16:46:13'),
    pd.to_datetime('2020-05-11 19:23:43'),
    pd.to_datetime('2024-04-20 12:09:27')  # Example date for the fourth cycle
]

# Add a column for cycle number
daily_df['Cycle'] = 0
for i, date in enumerate(halving_dates[:-1]):
    daily_df.loc[(daily_df['Timestamp'] >= date) & (daily_df['Timestamp'] < halving_dates[i+1]), 'Cycle'] = i + 1

# Analyze each cycle
for cycle in daily_df['Cycle'].unique():
    cycle_data = daily_df[daily_df['Cycle'] == cycle].copy()
    cycle_data.set_index('Timestamp', inplace=True)
    
    # Fit ARIMA model
    model = ARIMA(cycle_data['Close'], order=(1, 1, 1))  # Example order, adjust as needed
    model_fit = model.fit()
    
    # Print summary
    print(f"Cycle {cycle} ARIMA Model Summary:")
    print(model_fit.summary())
    
    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(cycle_data['Close'], label='Actual')
    plt.plot(model_fit.fittedvalues, color='red', label='Fitted')
    plt.title(f'Cycle {cycle} - ARIMA Model')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to plot Fear and Greed score with Bitcoin price for a specific halving cycle
def plot_cycle_fng_and_price(cycle_number, merged_df):
    cycle_data = merged_df[merged_df['Halving'] == cycle_number]
    days_from_start = (cycle_data['Timestamp'] - cycle_data['Timestamp'].iloc[0]).dt.total_seconds() / (24*60*60)
    
    fig, ax1 = plt.subplots(figsize=(15, 8))
    
    # Plot Bitcoin price
    ax1.plot(days_from_start, cycle_data['Close'], label='Price (USD)', color='blue', alpha=0.8)
    ax1.set_xlabel('Days from Start of Cycle')
    ax1.set_ylabel('Price (USD)', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')
    ax1.set_title(f'Fear and Greed Score with Bitcoin Price for Halving Cycle {cycle_number}')
    
    # Create a secondary y-axis for the Fear and Greed score
    ax2 = ax1.twinx()
    ax2.plot(days_from_start, cycle_data['value'].astype(float), label='Fear and Greed Score', color='orange', alpha=0.6)
    ax2.set_ylabel('Fear and Greed Score', color='orange')
    ax2.tick_params(axis='y', labelcolor='orange')
    
    # Add grid and legend
    ax1.grid(True)
    fig.tight_layout()
    plt.show()

# Plot for the 2nd halving cycle
plot_cycle_fng_and_price(2, merged_df)

# Plot for the 3rd halving cycle
plot_cycle_fng_and_price(3, merged_df)

# Plot for the 4th halving cycle
plot_cycle_fng_and_price(4, merged_df)
```
